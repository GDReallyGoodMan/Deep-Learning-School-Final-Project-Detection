```python
from ultralytics import YOLO
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("gdreallygoodman/superdataset")

print("Path to dataset files:", path)
```

```python
model_fast = YOLO('yolov8n.pt')
model_accurate = YOLO('yolov8m.pt')
```

```python
results_fast = model_fast.train(
    data='/kaggle/input/superdataset/data.yaml',
    epochs=50,
    imgsz=600,
    batch=128,
    device=[0, 1],
    workers=16,

    lr0=0.01,
    lrf=0.001,
    warmup_epochs=3,

    mosaic=1.0,
    mixup=0.1,
    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4,
    degrees=5.0,
    fliplr=0.5,
    translate=0.1,


    patience=20,
    save=True,
    save_period=10,
    plots=True,
    seed=42,
    verbose=True
)
```

```python
results_accurate = model_accurate.train(
    data='/kaggle/input/superdataset/data.yaml',
    epochs=80,
    imgsz=740,
    batch=50,
    device=[0, 1],
    workers=16,

    lr0=0.01,
    lrf=0.001,
    warmup_epochs=3,

    mosaic=1.0,
    mixup=0.1,
    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4,
    degrees=5.0,
    fliplr=0.5,
    translate=0.1,


    save=True,
    save_period=10,
    plots=True,
    seed=42,
    verbose=True,

    amp=True,
    cache=True 
)
```
